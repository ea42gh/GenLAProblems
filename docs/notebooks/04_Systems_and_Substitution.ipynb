# Systems and Substitution

Back/forward substitution and solution display.

```julia
using GenLAProblems
using Random
Random.seed!(41325);
```

## Build a system and reduce

```julia
A, X, B = gen_gj_pb(3, 5, 2; maxint=2, pivot_in_first_col=true, num_rhs=1, has_zeros=true)
pb = ShowGe{Rational{Int}}(A, B, tmp_dir="../../tmp")
ref!(pb; gj=true)
```

## Show system

```julia
show_system(pb; b_col=1)
```

## Back-substitution

```julia
show_backsubstitution!(pb, b_col=1, fig_scale=1.2)
```

## Forward-substitution (example)

```julia
A2 = [1 0 0; 2 1 0; -1 3 1]
b2 = [1, 2, 3]
show_forwardsubstitution(A2, b2; var_name="y")
```

## Solution vector

```julia
show_solution!(pb, b_col=1)
```
