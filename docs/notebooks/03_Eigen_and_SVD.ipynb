# Eigenproblems and SVD

Eigen and SVD table rendering with GenLAProblems.

```julia
using GenLAProblems
using Random
Random.seed!(2024);
```

## Eigen tables

```julia
A = [2 1 0; 1 2 0; 0 0 3]
h = nM.show_eig_tbl(A; fig_scale=1.2, mmLambda=11, color="blue", tmp_dir="../../tmp")
h
```

## SVD tables

```julia
B = [1 0 0; 0 2 0; 0 0 3]
h = nM.show_svd_table(B; fig_scale=1.2, mmS=7, mmLambda=10, tmp_dir="../../tmp")
h
```
