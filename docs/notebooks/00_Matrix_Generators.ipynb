# Matrix Generators

Examples of the matrix/problem generators available in GenLAProblems. These are adapted from `M337_Problems.ipynb` and `GenALProblems_examples.ipynb`.

```julia
using GenLAProblems, LaTeXStrings
using Random
Random.seed!(1234);
```

## Gaussian elimination problems

```julia
A, X, B = gen_gj_pb(3, 5, 2; maxint=3, pivot_in_first_col=false, num_rhs=2, has_zeros=false)
l_show( A, X, " = ", B)
```

## LU / PLU / LDLT

```julia
m, n, r = 4, 6, 3
pivot_cols, L, U, A = gen_lu_pb(m, n, r; maxint=3)
l_show( "A = L U: ", A, " = ", L, U)
```

```julia
pivot_cols, P, L, U, A = gen_plu_pb(4, 6, 3; maxint=3)
l_show( "A = P L U: ", A, " = ", P, L, U)
```

```julia
L, D, A = gen_ldlt_pb(4)
l_show( L"A = L D L^T : ", A, " = ", L, D, L')
```

## QR generators

```julia
A_qr = gen_qr_problem_3()
l_show( A_qr )
```

## Eigen / SVD generators

```julia
S,Λ,A=gen_symmetric_eigenproblem([1,1,0], maxint=2)
d,A_=factor_out_denominator(A)
#dS,S_=factor_out_denominator(S)
display(l_show(L"A = ", A, L",\quad S = ", S, L", \quad \Lambda = ", Λ))

h, _ = nM.show_eig_tbl(A_,Ascale=d,normal=true, fig_scale=1, mmLambda=11, color="blue", tmp_dir="../../tmp")
h
```

```julia
U,Σ,Vt,A = gen_svd_problem(3,4, [2,3],  maxint=2)
dA,intA=factor_out_denominator(A)
dU,intU=factor_out_denominator(U)
dVt,intVt=factor_out_denominator(Vt)

display( l_show( "A = ",A, " = ",U,Σ,Vt))

h, _ = nM.show_svd_table(intA, Ascale=dA, mmS=7, mmLambda=10, tmp_dir="../../tmp")
h
```
